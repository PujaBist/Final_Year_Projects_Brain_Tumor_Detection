**Evaulating all three model**


```python
import json
from src.config import METRICS_DIR

models = {
    'ResNet50'       : 'resnet50_test',
    'DenseNet121'    : 'densenet121_test',
    'EfficientNetB0' : 'efficientnetb0_test',
}

all_metrics = {}
for display_name, file_name in models.items():
    path = METRICS_DIR / f'{file_name}_metrics.json'
    with open(path, 'r') as f:
        all_metrics[display_name] = json.load(f)
    print(f"✅ Loaded: {file_name}")

print(f"\n{'='*72}")
print(f"  MODEL COMPARISON TABLE — BRAIN TUMOR CLASSIFICATION")
print(f"{'='*72}")
print(f"  {'Metric':<20} {'ResNet50':>14} {'DenseNet121':>13} {'EfficientB0':>13}")
print(f"  {'─'*65}")

rows = [
    ('Accuracy',    'accuracy'),
    ('Precision',   'precision'),
    ('Recall',      'recall'),
    ('F1 (macro)',  'f1_macro'),
    ('AUC-ROC',     'auc_macro'),
    ('MCC',         'mcc'),
    ('Kappa',       'kappa'),
]

for label, key in rows:
    row = f"  {label:<20}"
    for name in ['ResNet50', 'DenseNet121', 'EfficientNetB0']:
        val = all_metrics[name][key]
        if key in ['mcc', 'kappa']:
            row += f"  {val:>13.4f}"
        else:
            row += f"  {val*100:>12.2f}%"
    print(row)

print(f"{'='*72}")

best = max(all_metrics.items(), key=lambda x: x[1]['accuracy'])
print(f"\n  🏆 Best individual model : {best[0]}")
print(f"     Accuracy              : {best[1]['accuracy']*100:.2f}%")
print(f"     AUC-ROC               : {best[1]['auc_macro']*100:.2f}%")
print(f"     MCC                   : {best[1]['mcc']:.4f}")
print(f"\n  Next step → Ensemble all 3 models")
```

```python
# Cell 1 — load all 3 trained models
%matplotlib inline
from src.models import load_model
from src.ensemble import (
    get_all_probabilities,
    find_optimal_weights,
    evaluate_ensemble,
    compare_ensemble_vs_individuals,
    plot_confidence_distribution,
    weighted_soft_voting,
)

models = {
    'resnet50'      : load_model('resnet50'),
    'densenet121'   : load_model('densenet121'),
    'efficientnetb0': load_model('efficientnetb0'),
}

# Cell 2 — get probabilities from all 3 models
probs_dict, all_labels = get_all_probabilities(
    models, dataloaders['test']
)

# Cell 3 — find best weights
best_weights = find_optimal_weights(probs_dict, all_labels)

# Cell 4 — evaluate ensemble
ensemble_metrics = evaluate_ensemble(
    probs_dict, all_labels,
    best_weights, 'test'
)

# Cell 5 — compare all
import json
from src.config import METRICS_DIR

individual_metrics = {}
for name in ['resnet50', 'densenet121', 'efficientnetb0']:
    path = METRICS_DIR / f'{name}_test_metrics.json'
    with open(path) as f:
        individual_metrics[name] = json.load(f)

compare_ensemble_vs_individuals(
    ensemble_metrics,
    individual_metrics
)

# Cell 6 — confidence distribution
ensemble_probs = weighted_soft_voting(probs_dict, best_weights)
plot_confidence_distribution(probs_dict, ensemble_probs)
```
